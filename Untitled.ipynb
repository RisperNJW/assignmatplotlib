```python
import pandas as pd
data = pd.read_csv("kenya_primary_schools.csv")
data.head()
```

```python
data.isnull().sum()
data_clean = data.dropna()
data_clean.isnull().sum().sum()
```

```python
data_clean.describe()
data_clean.groupby('Province')['TotalEnrol'].mean().sort_values(ascending=False)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
sns.set(style="whitegrid")

data_clean['Index'] = range(len(data_clean))
sns.lineplot(data=data_clean, x='Index', y='TotalEnrol')
plt.title('Enrollment Trend')
plt.show()
```

```python
data_clean.groupby('Province')['TotalEnrol'].mean().sort_values().plot(kind='barh', color= ["red", "green", "yellow", "pink", "brown", "orange", "blue","black"])
plt.title('The Average Enrollment by Province')
plt.xlabel('Enrollment')
plt.show()
```

```python
sns.histplot(data_clean['PupilTeach'], bins=40, kde=True)
plt.title('Distribution of Pupil-Teacher Ratio')
plt.show()
```

```python
sns.scatterplot(data=data_clean, x='NoTeaching', y='TotalEnrol')
plt.title('Enrollment vs. Teaching Staff')
plt.show()
```

```python
data_clean.to_csv('kenya_primary_schools_clean.csv', index=False)
```
